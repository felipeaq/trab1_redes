```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("result.txt",header=None)
df[0]/=1000
df[1]*=1000
df.head()
```

```python
plt.figure(figsize=(18,10))

x=np.arange(0,400,5)
internet_to_h2=x*2+40
h1_to_h2=x*4
hand_shake_h1_h2=h1_to_h2
hand_shake_internet_h2=internet_to_h2
y=3*(2*h1_to_h2+internet_to_h2+hand_shake_h1_h2+hand_shake_internet_h2)
y=3*(2*4*x+2*x+40+4*x+2*x+40)
y=3*(8*x+2*x+4*x+2*x+80)
y=3*(16*x+80)
y=48*x+240
regressor = LinearRegression()  
regressor.fit(np.array(df[0][0:-1]).reshape(-1,1),np.array(df[1])[0:-1]) 
plt.plot(x,y,"-s",linewidth=10)
plt.plot(x,regressor.predict(x.reshape(-1,1)),"->",linewidth=3,color="black")
plt.plot(df[0],df[1],"-o",color="red")
plt.xlabel ("atraso (ms)")
plt.ylabel ("tempo total (ms)")
plt.yscale("linear")
plt.xscale("linear")
plt.legend(["48*x+240 (analtico)","{}*x+{} (regresso linear)".format(regressor.coef_[0],regressor.intercept_),"Simulado pelo mininet"])
```
